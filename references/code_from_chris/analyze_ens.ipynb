```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
idxes = {"u10":0, "v10":1, "t2m":2, "t850":3, "z500":4, "TP":5}
```

```python
# Open dataset
path = "../../data/model_output/2019_ens.h5"
ens = xr.open_dataset(path)
ens
```

```python
#Create subsets and renormalize
predicted = ens.forecast
truth = ens.ground_truth
```

```python
step = 2
var = 2
pert = 0

fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize = (25,10))
im1 = predicted.isel(ic = 0, step = step, var = var, pert = pert).plot(cmap = "RdBu_r", yincrease = False, ax = axs[0])
im1 = truth.isel(ic = 0, step = step, var = var).plot(cmap = "RdBu_r", yincrease = False, ax = axs[1])
```
