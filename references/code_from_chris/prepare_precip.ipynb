## Create yearly data for 2020-2022

```python
data_2020 = precip.isel(latitude = slice(0,720), time = slice(0,1464))
data_2020.to_netcdf(path+"out_of_sample/2020.nc")
```

```python
data_2021 = precip.isel(latitude = slice(0,720), time = slice(1464, 2924))
data_2021.to_netcdf(path+"out_of_sample/2021.nc")
```

```python
data_2022 = precip.isel(latitude = slice(0,720), time = slice(2924,4384))
data_2022.to_netcdf(path+"out_of_sample/2022.nc")
```
